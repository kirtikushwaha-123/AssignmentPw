1. What is a parameter?

   -> A parameter is a variable in a model that is learned from data during training.

    Example: The weights and biases in a neural network or coefficients in linear regression.

2. What is correlation?

    -> Correlation measures the strength and direction of a linear relationship between two variables.

    Values range from -1 (perfect negative) to +1 (perfect positive), with 0 meaning no linear relationship.

2.1 What does negative correlation mean?


    -> Negative correlation means that when one variable increases, the other tends to decrease.

    Example: Speed of a car ↑ → Travel time ↓.

3. Define Machine Learning. What are the main components in Machine Learning?


    -> Definition: Machine Learning (ML) is a field of AI that enables systems to learn patterns from data and make predictions/decisions without being explicitly programmed.

    Main components:

    1. Data
    2. Features (input variables)
    3. Model/Algorithm
    4. Loss function
    5. Optimization (training process)
    6. Evaluation (testing & metrics)

4. How does loss value help in determining whether the model is good or not?

    ->  The loss function measures how far the model’s predictions are from actual values.

    Lower loss → better performance.

    If loss is high, model predictions are poor.

5. What are continuous and categorical variables?

    ->  Continuous variables: Numeric values that can take any value in a range. (e.g., height, temperature).

    Categorical variables: Variables that represent categories or groups. (e.g., gender, country).

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

    -> Convert categories into numerical form.

    1. Label Encoding (assigns numbers to categories)

    2. One-Hot Encoding (creates binary columns)

    3. Ordinal Encoding (for ordered categories)

7. What do you mean by training and testing a dataset?

    -> Training dataset: Used to teach the model patterns in the data.

15. What is causation? Explain difference between correlation and causation with an example.

    Causation: One variable directly affects the other.

    Correlation: Variables move together but not necessarily because of each other.

     Example:

    Ice cream sales and drowning cases are correlated (both increase in summer).

    Causation is the hot weather, which causes both events independently.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

    Optimizer: An algorithm used to adjust model parameters (weights) to minimize loss.

    Types:

    Gradient Descent: Updates parameters in the direction of negative gradient.

    Stochastic Gradient Descent (SGD): Uses one sample at a time.

    Mini-batch Gradient Descent: Uses small batches of data.

    Adam: Combines momentum + adaptive learning rates (most popular).

    RMSprop: Adapts learning rate for each parameter.

     Example (Keras):

    from tensorflow.keras.optimizers import Adam
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

18. What does model.fit() do? What arguments must be given?

    -> Purpose:
      model.fit() is used to train a machine learning model on the given dataset.
      It learns the relationship between input features (X) and target labels (y).

      Arguments:

      X (features/input data): The independent variables (usually a NumPy array, Pandas DataFrame, or similar).

      y (labels/target): The dependent variable (what you want to predict).

      Example:
      model.fit(X_train, y_train)

19. What does model.predict() do? What arguments must be given?

    Purpose:
      After training, model.predict() is used to make predictions on new/unseen data.

    Arguments:

      X (input features only, no labels): The data you want predictions for.

    Example:

      y_pred = model.predict(X_test)

19. What are continuous and categorical variables?

    Continuous Variables:
      Numerical data that can take infinite values within a range.

      Examples: Height, Weight, Salary, Temperature.

      Usually represented as float/int values.

    Categorical Variables:

      Data that falls into distinct categories or groups.

      Examples: Gender (Male/Female), City (Delhi, Mumbai, Bangalore), Colors (Red, Green, Blue).

      Usually represented as strings or integer codes.

21. What is Feature Scaling? How does it help in Machine Learning?

    Feature Scaling:
      Process of bringing all features to a similar scale/range so that no feature dominates others.

    Why important:

      Many ML algorithms (like KNN, SVM, Gradient Descent, Neural Networks) are sensitive to feature magnitudes.

      Without scaling, features with larger ranges (like Salary in thousands) may dominate smaller ones (like Age in years).

22. How do we perform scaling in Python?

    Using scikit-learn (sklearn.preprocessing):

        Standardization (Z-score scaling):

      from sklearn.preprocessing import StandardScaler
      scaler = StandardScaler()
      X_scaled = scaler.fit_transform(X)


      (Mean = 0, Std Dev = 1)

      Normalization (Min-Max scaling):

      from sklearn.preprocessing import MinMaxScaler
      scaler = MinMaxScaler()
      X_scaled = scaler.fit_transform(X)

      (Values scaled between 0 and 1)

25. Explain Data Encoding

      Since ML models work with numbers, categorical data must be converted to numeric form.

      Label Encoding: Assigns each category a unique number.

      from sklearn.preprocessing import LabelEncoder
      encoder = LabelEncoder()
      y_encoded = encoder.fit_transform(y)


      One-Hot Encoding: Creates binary (0/1) columns for each category.

      from sklearn.preprocessing import OneHotEncoder
      import pandas as pd

      encoder = OneHotEncoder()
      X_encoded = encoder.fit_transform(X[['city']]).toarray()


      Example:
        Column "City" = [Delhi, Mumbai, Delhi]

        Label Encoding → [0, 1, 0]

        One-Hot Encoding →

            Delhi  Mumbai
                  [1,0]
                  [0,1]
                  [1,0]
