4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [85, 89, 90, 92, 88],
    'Science': [78, 75, 92, 88, 90],
    'English': [90, 88, 85, 87, 89],
    'History': [70, 75, 80, 72, 78]
}

df = pd.DataFrame(data)

# 1. Calculate the correlation matrix
corr_matrix = df.corr()

# 2. Visualize using Seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")
plt.show()
```

8. Create a line plot with multiple lines using MatplotlibA


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 4, 3, 2, 1]

# Create the plot
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3', color='red', linestyle=':', marker='^')

# Add title and labels
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

9. Generate a Pandas DataFrame and filter rows where a column value is greater than a thresholdA

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [85, 42, 77, 90, 65]
}

df = pd.DataFrame(data)

# Define a threshold
threshold = 70

# Filter rows where 'Score' is greater than threshold
filtered_df = df[df['Score'] > threshold]

print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (Score > 70):")
print(filtered_df)
```

10. Create a histogram using Seaborn to visualize a distributionA

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (normally distributed)
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data, bins=30, kde=True, color='skyblue')

# Add labels and title
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

11. Perform matrix multiplication using NumPyA


```python
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.dot(A, B)

# Or using the @ operator (Python 3.5+)
# result = A @ B

# Print the result
print("Matrix Multiplication Result:\n", result)
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate example data
np.random.seed(42)
df = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'z': np.random.rand(50),
    'color': np.random.choice(['A', 'B', 'C'], size=50)
})

# Create 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='color',
                    title='3D Scatter Plot Example')

fig.show()
```
