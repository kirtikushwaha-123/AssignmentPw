Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

  -> Heteroscedasticity refers to a situation where the variance of the residuals (errors) is not constant across all levels of the independent variable x. In other words, the spread of errors changes depending on the value of x.

  Why it‚Äôs important:

    It violates the assumption of homoscedasticity in linear regression.

    It leads to inefficient and biased estimates of standard errors, which can affect hypothesis testing and confidence intervals.

    It can cause the model to underestimate or overestimate the uncertainty, resulting in incorrect conclusions.

Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values using the model
Y_pred = model.predict(X)

# Plotting the data points and the regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

Question 8: Implement polynomial regression on the following data:
‚óè X = [1, 2, 3, 4, 5]
3
‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform input data to include polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict Y values using the model
Y_pred = model.predict(X_poly)

# Plotting the original data and the polynomial curve
plt.scatter(X, Y, color='blue', label='Data points')

# Generate a smooth curve for plotting
X_curve = np.linspace(1, 5, 100).reshape(-1, 1)
X_curve_poly = poly.transform(X_curve)
Y_curve = model.predict(X_curve_poly)

plt.plot(X_curve, Y_curve, color='red', label='Polynomial fit (degree=2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (2nd Degree)')
plt.legend()
plt.grid(True)
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:
‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot the residuals
plt.scatter(X, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```

model.



1. Handling Multicollinearity

Multicollinearity occurs when two or more independent variables are highly correlated, making it difficult to distinguish their individual effects.

Steps to address it:

- > Check correlation matrix or VIF values:

 - > Remove or combine correlated features:

 - > Apply dimensionality reduction:

      Techniques like Principal Component Analysis (PCA) can transform correlated features into uncorrelated components.

 - > Use regularization techniques:

      Models like Ridge Regression or Lasso Regression add penalties to coefficients, reducing their impact and managing multicollinearity.

 - > Feature engineering:

      Create new variables that capture the relationship more meaningfully without redundancy.



2. Handling Heteroscedasticity

Heteroscedasticity happens when the variance of errors is not constant across observations, affecting the reliability of confidence intervals and significance tests.

Steps to address it:

 - > Visualize residuals:

Plot residuals to confirm if the variance changes with fitted values or predictors.

 - > Transform the dependent variable:

Applying transformations like logarithmic, square root, or Box-Cox can stabilize variance.

 - > Use weighted regression:

Assign weights inversely proportional to the variance, giving less weight to observations with higher variance.

 - > Use robust standard errors:

Adjust standard errors to account for heteroscedasticity without transforming the model structure.

 - > Model selection:

Consider models like Generalized Least Squares (GLS) that explicitly handle changing variance.
